```python
import numpy as np
from numpy.conftest import dtype
```

```python
# 特殊矩阵的构造
# 单位矩阵
arr=np.eye(4,dtype=int)
print(arr)
```

```python
# 对角矩阵
arr=np.diag([1,2,3,5])
print(arr)
```

```python
# 随机数组的生成
# 生成 [0,1) 之间的随机浮点数（均匀分布）
arr=np.random.rand(2,3)
# 参数：行，列
print(arr)
```

```python
# 生成指定范围区间的随机浮点数
# [l,r) shape
arr=np.random.uniform(3,5,(2,3))
print(arr)
```

```python
# 生成指定范围区间的随机整数
# [l,r) shape
arr=np.random.randint(3,10,(2,3))
print(arr)
```

```python
# 生成随机数列（正态分布）
# 生成标准正态分布随机数（均值为0，方差为1）
# 两边概率小，中间概率大
arr=np.random.randn(2,3)
print(arr)
```

```python
# 设置随机种子
# 每次运行生成的都是一样的
np.random.seed(10)
arr=np.random.randint(1,10,(2,5))
print(arr)
```
