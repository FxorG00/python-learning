## 2.2. ndarray
### 2.2.1 ndarray 的特性（多维性）

```python
import numpy as np
```

```python
arr=np.array(5) # 创建 0 维的 ndarray 数组
print(f"维度：{arr.ndim}")
```

```python
arr=np.array([1,2,3]) # 多一对中括号，加一维
print(arr)
print(arr.ndim)
```

```python
arr=np.array([[1,2,3],[4,5,6]]) # 创建一个矩阵，填入每一行
# 2*3 两行，三列
print(arr)
print(arr.ndim)
```

#### 同质性：所有元素类型一致

```python
arr=np.array([1,23,'hello'])
# 1,23 被转化成字符串了
print(arr)
print(arr.ndim)
```

```python
arr=np.array([1,2.5])
# 把 1 转化成浮点数
# 如果数组中有小数，dtype 会自动变成 float64（避免精度丢失）
print(arr)
```

### 2.2.2. ndarray 的属性

#### shape：输出数组的形状，多少行多少列

```python
arr=np.array([[1,2,3],[4,5,6]]) # 创建一个矩阵，填入每一行
# 2*3 两行，三列
print(arr)
print(arr.shape)
```

#### ndim：数组的维度

#### size： 数组中一共有多少个元素

#### dtype：数组元素的类型，int64（默认）
如果数组中有小数，dtype 会自动变成 float64（避免精度丢失）
