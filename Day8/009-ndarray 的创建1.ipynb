```python
import numpy as np
```

```python
# 基础
# np.array() 是一个方法，传入一个 list
arr=np.array([1,2,3],dtype=np.float64)
# dtype 用于指定元素的类型
print(arr)
print(arr.ndim) # ndim 是属性
```

```python
# copy
arr1=np.copy(arr)
# 新建一个arr1，它的内容从 arr copy 过来
print(arr1)
```

```python
# 预定义形状
# 全 0
arr=np.zeros((2,3),dtype=np.int64)
print(arr)
arr=np.zeros((2,)) # 创建的是一个一维的向量
print(arr)
```

```python
# 全 1
arr=np.ones((2,3))
print(arr)
```

```python
# 未初始化
arr=np.empty((4,2))
print(arr)
```

```python
# 填充
arr=np.full((4,2),fill_value=5)
print(arr)
```

```python
arr2=np.zeros_like(arr)
print(arr2)
# arr2 跟 arr 形状一样，但是用 0 填充
```
